# Zomi Dataset Explorer
Load, explore, and search the Zomi corpus.


```python
from pathlib import Path
import random

DATA = Path("../data")
FILES = list(DATA.glob("zomi_clean_p*.txt"))
print(f"Found {len(FILES)} corpus files")
```

```python
all_lines = []
for f in FILES:
    lines = [l.strip() for l in f.read_text(encoding="utf-8", errors="ignore").split("\n") if l.strip()]
    all_lines.extend(lines)
    print(f"{f.name}: {len(lines)} lines")
print(f"\nTotal: {len(all_lines):,} lines")
```

```python
print("Sample sentences:\n")
for i in range(5):
    print(f"{i+1}. {random.choice(all_lines)}")
    print()
```

```python
# Search for a word or phrase
keyword = "gamtang"  # Change this to search for any Zomi word
matches = [l for l in all_lines if keyword.lower() in l.lower()]
print(f"Found {len(matches)} lines containing '{keyword}'")
for m in matches[:5]:
    print(f"  • {m[:120]}...")
```

```python
# Word frequency (top 20)
from collections import Counter
words = " ".join(all_lines).lower().split()
common = Counter(words).most_common(20)
print("Top 20 words:\n")
for word, count in common:
    print(f"  {word:>12}: {count:>6,}")
```
